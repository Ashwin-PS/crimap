```
from google.colab import files
import pandas as pd
from difflib import SequenceMatcher

def similar(word,dictionary):
    for key in dictionary.keys():
      if SequenceMatcher(None, word, key.split(' ')[0]).ratio() >= 0.8:
        return key
    return '0'
  
  
def similar2(word,words):
    for key in words:
      if SequenceMatcher(None, word, key.split(' ')[0]).ratio() >= 0.8:
        return key
    return False
```

```
#Processing first version of dataset to remove '\n' and '\r' from keywords

data = pd.read_csv('dataset4.csv')
column = 'Keywords'
i = 0
for keywords in data[column]:
  newKeyword = str(keywords).replace('.','').replace('\r','').replace('\n','').replace(',','').replace('"','').replace(':','')
  data.set_value(i,column,newKeyword)
  i += 1
  
data.to_csv('Dataset4.csv',index=False)
files.download('Dataset4.csv')
```

```
count = dict()
for keywords in data[column]:
  for keyword in str(keywords).split(' '):
    if keyword in count.keys():
      count[keyword] += 1    
    else:
      similarKeyword = similar(keyword,count)
      if similarKeyword in count.keys():
        if keyword in similarKeyword.split(' '):
          count[similarKeyword] += 1
        else:
          count[similarKeyword + ' ' + keyword] = count[similarKeyword] + 1          
          del count[similarKeyword]  
      else :
        count[keyword] = 1
      
      
      
```

```
def findCategory(title,keywords):
  if title != 'nan':
    for i,row in titleCategories.iterrows():
      if similar2(title,str(row['Keyword']).split(' ')):
        return row['Category']
  for i,row in keywordCategories.iterrows():
    for keyword in keywords.split(' '):
      if similar2(keyword,str(row['Keyword']).split(' ')):
        return row['Category']
    
```

```
Category = []
for i,row in data.iterrows():
  Category += [findCategory(str(row['Title']),str(row['Keyword']))]
```

```
data['Category'] = Category
data.to_csv('Dataset3.csv',index=False)
files.download('Dataset3.csv')
```
